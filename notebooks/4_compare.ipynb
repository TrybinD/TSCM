```python
import pandas as pd
```

```python
solvers = ["random", "greedy", "mip_paper", "mip", "dp"]
```

```python
full_df_results = []

for solver in solvers:
    df = pd.read_csv(f"../data/solutions/{solver}.csv", index_col=0)

    df.columns = pd.MultiIndex.from_tuples([(solver, col) for col in df.columns])

    full_df_results.append(df)

full_df_results = pd.concat(full_df_results, axis=1)
```

```python
full_df_results
```

```python
full_df_results[("mip_paper", "cost")] >= full_df_results[("mip", "cost")]
```

```python
full_df_results[("mip", "cost")] >= full_df_results[("dp", "cost")]
```

```python
full_df_results[("mip", "cost")] == full_df_results[("dp", "cost")]
```

```python
full_df_results[("mip", "solution")] == full_df_results[("dp", "solution")]
```
